# 학습, 대략적인 알고리즘 성능 체크

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.iloc[:,1:-1], data.iloc[:,-1])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
print(lr.score(X_train, y_train))
print(lr.score(X_test, y_test))
```

```python
tree = DecisionTreeRegressor()
tree.fit(X_train, y_train)
print(tree.score(X_train, y_train))
print(tree.score(X_test, y_test))
```

```python
forest = RandomForestRegressor()
forest.fit(X_train, y_train)
print(forest.score(X_train, y_train))
print(forest.score(X_test, y_test))
```

```python
knn = KNeighborsRegressor()
knn.fit
```

# cross validation 돌려보기 알고리즘 대략적 평가

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(lr, data.iloc[:,1:-1], data.iloc[:,-1], cv=10).mean()
```

```python
cross_val_score(tree, data.iloc[:,1:-1], data.iloc[:,-1], cv=10).mean()
```

```python
cross_val_score(forest, data.iloc[:,1:-1], data.iloc[:,-1], cv=10).mean()
```

```python
cross_val_score(knn, data.iloc[:,1:-1], data.iloc[:,-1], cv=10).mean()
```
