### cross_validation 을 이용한 알고리즘 선정

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_score_list = []
```

```python
cross_score_list.append(cross_val_score(RandomForestRegressor(), data2.iloc[:,:-1], data2.iloc[:,-1], cv=10).mean())
cross_score_list.append(cross_val_score(LinearRegression(), data2.iloc[:,:-1], data2.iloc[:,-1], cv=10).mean())
cross_score_list.append(cross_val_score(DecisionTreeRegressor(), data2.iloc[:,:-1], data2.iloc[:,-1], cv=10).mean())
cross_score_list.append(cross_val_score(Lasso(), data2.iloc[:,:-1], data2.iloc[:,-1], cv=10).mean())
cross_score_list.append(cross_val_score(Ridge(), data2.iloc[:,:-1], data2.iloc[:,-1], cv=10).mean())
cross_score_list.append(cross_val_score(SVR(), data2.iloc[:,:-1], data2.iloc[:,-1], cv=10).mean())
poly = make_pipeline(PolynomialFeatures(2), LinearRegression())
cross_score_list.append(cross_val_score(poly, data2.iloc[:,:-1], data2.iloc[:,-1], cv=10).mean())
rp2 = make_pipeline(PolynomialFeatures(2), Ridge(alpha=0.1))
cross_score_list.append(cross_val_score(rp2, data2.iloc[:,:-1], data2.iloc[:,-1], cv=10).mean())
```

```python
cross_score_list
```

따라서 대략적으로 RandomForestRegressor()로 선택.

### 버블 여부와의 상관관계

```python
data_buble = data.iloc[:,1:]
```

```python
data_buble['buble'] = data_buble['price'] - forest_real.predict(data_buble.iloc[:,:-1])
```

```python
data_buble.head()
```

```python
# 실제 - 예측 = 양수 => 버블이다. 사지마라. 0
# 실제 - 예측 = 음수 => 안버블이다. 사라. 2
# 버블 0
# 적정가 1
# 안버블 2
```

```python
for i in range(0,2963):
    if data_buble.iloc[i,-1] > (data_buble.iloc[i,-2] * 0.01) :
        data_buble.iloc[i,-1] = 0
    elif data_buble.iloc[i,-1] < -(data_buble.iloc[i,-2] * 0.01) :
        data_buble.iloc[i,-1] = 2
    else:
        data_buble.iloc[i,-1] = 1
```

```python
data_buble.buble.value_counts()
```

```python
data_buble.to_csv('buble_real.csv',index =False)
```
